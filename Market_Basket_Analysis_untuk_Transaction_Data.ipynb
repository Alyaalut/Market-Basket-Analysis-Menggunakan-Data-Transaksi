```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#membaca file csv
data = pd.read_csv('transaction_data.csv', engine='python')

#menampilkan beberapa baris pertama
print(data.head())
```

perlu mengetahui kondisi data, dengan memeriksa data yang kosong.

```python
# Bersihkan data dari nilai yang hilang atau tidak konsisten
data.info()
```

```python
# mengecek nilai NaN atau NULL
data.isnull().sum()
```

```python
# menghapus baris yang mengandung nilai NaN atau NULL
data.dropna(inplace=True)
data.info()
```

Setelah menghilangkan data hilang, sekarang data menjadi 1080910 entri. Maka bisa dikatakan setidaknya terdapat satu nilai yang hilang. Untuk mengidentifikasi pola pembelian barang oleh pelanggan dan merumuskan stategi pemasaran berdasarkan aturan asosiasi yang dihasilkan. Pertama perlu membuat agregat berdasarkan TransactionId untuk mendapatkan daftar barang per transaks

```python
#agrerat TransactionId

basketplus = data.groupby(['TransactionId','ItemDescription'])['NumberOfItemsPurchased'].sum().unstack().reset_index().fillna(0).set_index('TransactionId')
basketplus
```

Dalam mengidentifikasi pola pembelian barang oleh pelanggan, penting mengetahui suatu barang terbeli atau tidak. Sebab, kita hanya ingin mengetahui, apa hubungan antara membeli beberapa barang dan membeli beberapa barang lainnya. Jadi, kita perlu mengkodekan data keranjang menjadi data biner yang menunjukkan apakah suatu barang dibeli (1) atau tidak (0).

```python
# Kode Beli Barang

def encode_units(x):
  if x<=0:
    return 0
  if x>= 1:
    return 1

basket_encode_plus = basketplus.applymap(encode_units)
basket_encode_plus
```

Kemudian perlu mengetahui hubungan antara 2 atau lebih item yang dibeli berdarkan data yang ada.  Jadi, kurang bermanfaat jika transaksi hanya membeli satu item saja.  Oleh karena itu, langkah selanjutnya adalah menyaring transaksi pembelian lebih dari 1 item.

```python
#Filter

basket_filter_plus = basket_encode_plus[(basket_encode_plus > 0).sum(axis=1) >= 2]
basket_filter_plus
```

Berdasarkan hasil di atas terlihat terdapat 18.338 transaksi yang membeli lebih dari satu item. Setelah menghasilkan dataset tersebut, tahap selanjutnya adalah menerapkan algoritma Apriori. Algoritma Apriori digunakan untuk menemukan frequent itemset, yaitu item atau kombinasi item yang sering muncul secara bersamaan dalam transaksi. Pada penelitian ini, nilai minimum support (min_support) ditetapkan sebesar 0,03, sehingga hanya item atau kombinasi item yang muncul pada minimal 3% dari keseluruhan transaksi yang akan dianggap sebagai frequent itemset. Penentuan nilai tersebut bertujuan untuk menyaring item yang cukup sering muncul sehingga pola yang dihasilkan lebih representatif dan tidak menghasilkan terlalu banyak maupun terlalu sedikit kombinasi item.

```python
#Apriori

from mlxtend.frequent_patterns import apriori

frequent_itemsets_plus = apriori(basket_filter_plus, min_support=0.03,
                                 use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

frequent_itemsets_plus['length'] = frequent_itemsets_plus['itemsets'].apply(lambda x: len(x))

frequent_itemsets_plus
```

Algoritma Apriori digunakan untuk menemukan frequent itemset, yaitu item atau kombinasi item yang sering muncul secara bersamaan dalam transaksi berdasarkan nilai minimum support yang telah ditentukan.

Length = 1 → menunjukkan 1 item yang sering dibeli.
Contoh: WHITE HANGING HEART T-LIGHT HOLDER memiliki support 0,122205, artinya produk tersebut muncul pada sekitar 12,22% dari seluruh transaksi.

Length = 2 → menunjukkan kombinasi 2 item yang sering dibeli secara bersamaan.
Contoh: LUNCH BAG SUKI DESIGN dan LUNCH BAG RED RETROSPOT muncul bersama pada sekitar 3,03% transaksi.


Setelah menerapkan algoritma apriori dan menemukan barang yang sering dibeli, kini dapat menerapkan aturan asosiasi. Dari aturan asosiasi, kita bisa mengetahui informasi bahkan menemukan pengetahuan tentang barang mana yang lebih efektif untuk dijual bersama.


```python
# Asosiasi
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets_plus, metric='lift',
                  min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)
```

Asosiasi adalah aturan yang menunjukkan hubungan antaritem dalam transaksi, yaitu jika pelanggan membeli suatu item (antecedent), maka ada kecenderungan pelanggan juga membeli item lain (consequent).

Dari hasil asosiasi diperoleh bahwa item Green Regency Teacup And Saucer dan Pink Regency Teacup And Saucer sangat bagus untuk dijual secara bersamaan. Karena pada dua item tersebut memiliki asosiasi paling tinggi dengan nilai lift yang paling tinggi, yaitu sebesar 15.028131. Semakin tinggi nilai lift, semakin tinggi pula hubungan antar item. Jika nilai lift lebih dari satu, maka cukup dikatakan bahwa kedua item tersebut saling berhubungan. Kemudian dari hasil asosiasi item Green Regency Teacup And Saucer dan Pink Regency Teacup And Saucer diperoleh nilai support sebesar 0.34518. Dapat diartikan terdapat 3.4518% dari total transaksi yang menjual kedua item tersebut secara bersamaan.
Perlu diketahui bahwa jika antesedennya lebih tinggi dari konsekuennya, maka aturan yang diterapkan adalah aturan nomor 1 dan sebaliknya. Dalam hal ini, nilai antesedennya lebih tinggi dibandingkan nilai konsekuensinya. Artinya dapat menerapkan aturan nomor 1 yaitu Green Regency Teacup And Saucer → Pink Regency Teacup And Saucer. Dapat dijelaskan bahwa pelanggan akan cenderung membeli Pink Regency Teacup And Saucer setelah membeli Green Regency Teacup And Saucer.
